```python
from numpy import *
from scipy.optimize import fsolve
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
import math as math
```

Here we will solve the equation itself then look at the equilibrium positions for each epsilon defined in the latex

```python
def ODE1A(theta,t,g,a,w):
    theta1 = theta[0]
    theta2 = theta[1]
    dtheta1 = theta2
    dtheta2 = -(g/a)*np.sin(theta1)+w**2*np.cos(theta1)*np.sin(theta1)
    dtheta = [dtheta1,dtheta2]
    return dtheta
```

```python
a = 1
g = 9.81
theta_0 = [1,0]
w = 1
```

```python
t = np.linspace(0,10,150)
```

```python
theta = odeint(ODE1A,theta_0,t,args=(g,a,w))
```

```python
plt.figure()
plt.plot(t,theta[:,0],'b--',label = r'Random Initial Conditions')
plt.ylabel(r'$\theta$')
plt.xlabel('time')
plt.title('Angle From the Downwards Vertical')
plt.legend(loc='best')
plt.show()
```

Now let us look at the cases where the square of the angular velocity of the wire is greater than the acceleration due to gravity, and the case where it is less than the acceleration due to gravity. 

```python
w_1 = 2*g
w_2 = g
w_3 = g/2
```

```python
theta1 = odeint(ODE1A,theta_0,t,args=(g,a,w_1))
theta2 = odeint(ODE1A,theta_0,t,args=(g,a,w_2))
theta3 = odeint(ODE1A,theta_0,t,args=(g,a,w_3))
```

```python
plt.figure()
plt.plot(t,theta1[:,0]+2,'b--',label = r'$a\Omega>g$')
plt.plot(t,theta2[:,0],'k',label = r'$a\Omega=g$')
plt.plot(t,theta3[:,0]-2,'r--',label = r'$a\Omega<g$')
plt.ylabel(r'$\theta$')
plt.xlabel('time')
plt.title('Angle From the Downwards Vertical for Different Angular Velocities')
plt.legend(loc='best')
plt.show()
```

I offset the graphs so they wouldn't stack on top of each other. But we see here that the particle oscillates with a larger frequency when $a\Omega$ is large compared to $g$, similarly, when $a\Omega$ is smaller than $g$ the frequency decreases compared to their equality.
