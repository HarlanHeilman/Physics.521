```python
from numpy import *
from scipy.optimize import fsolve
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
```

We will let the driving "force" be some quadratic function of time and $\phi$ is the angle for the driven motion

```python
def ODE1A(theta,t,g,a):
    theta1 = theta[0]
    theta2 = theta[1]
    dtheta1 = theta2
    dtheta2 = -(g/a)*np.sin(theta1)
    dtheta = [dtheta1,dtheta2]
    return dtheta
```

```python
def ODE1B(theta,t,g,a):
    phi1 = theta[0]
    phi2 = theta[1]
    dphi1 = phi2
    dphi2 = -(g/a)*np.sin(phi1)+.05*t**2
    dphi= [dphi1,dphi2]
    return dphi
```

```python
def ODE2A(theta,t,g,a):
    psi1 = theta[0]
    psi2 = theta[1]
    dpsi1 = psi2
    dpsi2 = -(g/a)*psi1
    dpsi= [dpsi1,dpsi2]
    return dpsi
```

```python
a = 1
g = 9.81
theta_0 = [2,0]
```

```python
t = np.linspace(0,10,150)
```

```python
theta = odeint(ODE1A,theta_0,t,args=(g,a))
phi = odeint(ODE1B,theta_0,t,args=(g,a))
psi = odeint(ODE2A,theta_0,t,args=(g,a))
```

```python
plt.figure()
plt.plot(t,theta[:,0],'b--',label = r'Non Driven')
plt.plot(t,phi[:,0],'r-',label = r'Driven')
plt.plot(t,psi[:,0],'-',label = r'Small Angle')
plt.legend(loc='best')
plt.show()
```
