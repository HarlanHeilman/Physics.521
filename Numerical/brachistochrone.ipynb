```python
from numpy import *
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
import numpy as np
```

```python
x0 = .5
y0 = .5
```

```python
def myFunction (z):
    a = z[0]
    t = z[1]
    
    F = empty((2))
    F[0] = x0 + a*(t - np.sin(t))
    F[1] = y0 + a*(1 - np.cos(t))
    return F
```

```python
zGuess = array([1,1])
```

```python
z = fsolve(myFunction,zGuess)
print(z)
```

```python
A = z[0]
theta = np.linspace(0,z[1],100)
```

```python
x = x0 + A*(theta - np.sin(theta))
y = y0 + A*(1 - np.cos(theta))
```

```python
f = plt.figure()
plt.plot(x,y)
plt.title('Brachistochrone for $x_0 = y_0 = .5$m', fontsize=11)
plt.xlabel('$y$(m)')
plt.ylabel('$x$(m)')
plt.show()
```

```python
f.savefig("brachistochrone.pdf", bbox_inches = 'tight')
```
